 group function

```python

df['Country'].value_counts()
```

```python
country_df=df.groupby(['Country'])
```

```python
country_df.get_group('Canada').head(5)
```

```python
filt = df['Country'] == 'India'
df.loc[filt]['SocialMedia'].value_counts()
```

```python
country_df['SocialMedia'].value_counts().loc['India']
```

```python
country_df['ConvertedComp'].agg(['median', 'mean'])
```

```python
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```
