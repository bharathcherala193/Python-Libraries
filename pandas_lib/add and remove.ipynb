## Add/Remove Rows and Columns from DataFrames

```python
import pandas as pd
```

```python
people = {
    "first": ["Eren", "Iron", "Rob"],
    "last": ["yeager", "Man", "Roe"],
    "email": ["erenyeager@gmail.com", "ironman@email.com", "robroe@email.com"]
}
```

```python

df = pd.DataFrame(people)
```

```python
df
```

```python
df['first'] + ' ' + df['last']
```

```python
df['full name']=df['first'] + ' ' + df['last']
```

```python
df
```

To delete columns:

```python
df.drop(columns=['last'])
```

```python
df.drop(columns=['first', 'last'], inplace=True)
```

```python
df
```

```python
df[['first', 'last']] = df['full name'].str.split(' ', expand=True)
```

```python
df
```

```python
df.drop(index=1,inplace=True)
```

```python
df.drop(index=df[df['last'] == 'Roe'].index)
```
