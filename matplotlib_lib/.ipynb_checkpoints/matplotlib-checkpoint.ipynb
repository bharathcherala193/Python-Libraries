## Mathplotlib

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel('dataset.xlsx')
```

```python
df.head(4)
```

```python
df.info()
```

 # BAR PLOT
 -- bar to reprasent the data categories ,with bar lenght and height propotional to their values

```python
# 1.Plot the number of records for each Microscope_Type.
Microscope_counts =df['Microscope_Type'].value_counts()
plt.figure()
plt.bar(Microscope_counts.index,Microscope_counts.values)
plt.xlabel('Microscope_type')
plt.ylabel('counts')
plt.title('Microscope_counts')
plt.show()
```

```python
# 2.Plot average Embryologist Experience for each Embryologist_Id
avg = df.groupby('Embryologist_ID')['Embryologist_Experience_Years'].mean()

plt.figure()
plt.bar(avg.index, avg.values)
plt.xlabel("Embryologist ID")
plt.ylabel("Average Experience (Years)")
plt.title("Average Experience per Embryologist")
plt.show()
```

# PIE CHART
-- is to create circle stastical graphical repasentaion divided into slices to percentage of single value

```python
# 1.Show percentage of Fertilization Success (0 vs 1)
fert= df['Fertilization_Success'].value_counts()

plt.figure()
plt.pie(fert.values, labels=fert.index, autopct='%1.1f%%')
plt.title("Fertilization  Distribution")
plt.show()
```

```python
# Show distribution of Usable_Embryo (0 vs 1)

usable = df['Usable_Embryo'].value_counts()

plt.figure()
plt.pie(usable.values, labels=usable.index, autopct='%1.2f%%')
plt.title("Usable Embryo Distribution")
plt.show()
```

# LINE PLOT
-- connect data points with staright lines

```python
# Plot Lab Temperature over time (Selection_Date)

df_sorted = df.sort_values("Selection_Date")

plt.figure()
plt.plot(df_sorted['Selection_Date'], df_sorted['Lab_Temperature_C'])
plt.xlabel("Selection Date")
plt.ylabel("Lab Temperature")
plt.title("Lab Temperature Over Time")
plt.xticks(rotation=45)
plt.show()
```

# HISTOGRAM
-- it visualize the distribution of numerical data

```python
# Plot distribution of Total_Motility_Percent
plt.figure()
plt.hist(df['Total_Motility_Percent'], bins=10)
plt.xlabel("Total  Percent")
plt.ylabel("Frequency")
plt.title("Distribution of Total Motility")
plt.show()
```

# SCATTER PLOT
-- it visualize the relationship b/w two numet=rical var


```python
# Scatter plot between Sperm Concentration and Total Motility.
plt.figure()
plt.scatter(df['Sperm_Concentration_M_per_ml'], df['Total_Motility_Percent'])
plt.xlabel("Sperm Concentration")
plt.ylabel("Total Motility Percent")
plt.title("Concentration vs Motility")
plt.show()
```

```python
# Check relationship between Embryologist Experience and Fertilization Success
plt.figure()
plt.scatter(df['Embryologist_Experience_Years'], df['Fertilization_Success'])
plt.xlabel("Experience ")
plt.ylabel("Fertilization Success")
plt.title("Experience vs Fertilization Success")
plt.show()
```

# BOXPLOT
    -- is a graph used to show the distribution of numerical data using five important values:
    -Minimum
    -First Quartile (Q1) â€“ 25%
    -Median (Q2) â€“ 50%
    -Third Quartile (Q3) â€“ 75%-
    -M-aximum

```python
plt.figure()
plt.boxplot(df['Sperm_Concentration_M_per_ml'])
plt.ylabel("Sperm Concentration (M/ml)")
plt.title("Boxplot of Sperm Concentration")
plt.show()
```

```python
group0 = df[df['Fertilization_Success'] == 0]['Total_Motility_Percent']
group1 = df[df['Fertilization_Success'] == 1]['Total_Motility_Percent']

plt.figure()
plt.boxplot([group0, group1])
plt.xticks([1, 2], ['Failure', 'Success'])
plt.ylabel("Total Motility Percent")
plt.title("Motility vs Fertilization Success")
plt.show()
```
